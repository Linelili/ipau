# Now include the model

```python
model = cobra.io.read_sbml_model('iPAU.xml')
print(len(model.metabolites))
```

```python
charge_dict = {}
formula_dict = {}

for met in model.metabolites:
    query_string = "SELECT formula FROM modelseed_compounds WHERE comp_id = '" + met.id[:-2]+ "';"
    result =  pd.read_sql_query(query_string, cnx)
    if result.empty != True:
        #print(met, result)
        formula_dict[met.id] = result.loc[0, 'formula']
    #print(result.loc[0, 'formula'])
    
    query_string = "SELECT charge FROM modelseed_compounds WHERE comp_id = '" + met.id[:-2]+ "';"
    result =  pd.read_sql_query(query_string, cnx)
    if result.empty != True:
        #print(met, result)
        #formula[met.id] = result.loc[0, 'formula']
        charge_dict[met.id] = result.loc[0, 'charge']
    
print(len(charge_dict))
print(len(formula_dict))
```

```python
### with 'old' charges and formulas
reactions_old = []
charge_old = []
mass_old = []
for react in model.reactions:
    result = react.check_mass_balance()
    if len(result) == 0:
        continue
    else: 
        reactions_old.append(react.id)
        masses = {}
        for key in result.keys():
            if key == 'charge':
                charge_old.append(result['charge'])
            else:
                masses[key] = result[key]
        if 'charge' not in result.keys():
            charge_old.append('')
        mass_old.append(masses)
print(len(reactions_old), len(charge_old), len(mass_old))
```

```python
diff_charge = []
diff_formula = []
for met in model.metabolites: 
    if met.id in charge_dict:
        if met.charge != int(charge_dict[met.id]):
            diff_charge.append(met.id)
            met.charge = int(charge_dict[met.id])
        if met.formula != formula_dict[met.id]:
            diff_formula.append(met.id)
            met.formula = formula_dict[met.id]
print(len(diff_charge))
print(len(diff_formula))
```

```python
### with 'new' charges and formulas
reactions = []
charge = []
mass = []
for react in model.reactions:
    result = react.check_mass_balance()
    if len(result) == 0:
        continue
    else: 
        reactions.append(react.id)
        masses = {}
        for key in result.keys():
            if key == 'charge':
                charge.append(result['charge'])
            else:
                masses[key] = result[key]
        if 'charge' not in result.keys():
            charge.append('')
        mass.append(masses)
print(len(reactions), len(charge), len(mass))
```

```python
intersec = [react for react in reactions_old if react in reactions]
print(len(intersec))
```

```python
### with 'new' charges and formulas
reactions_intersec = []
charge_intersec = []
mass_intersec = []
for react in intersec:
    result = model.reactions.get_by_id(react).check_mass_balance()
    if len(result) == 0:
        continue
    else: 
        reactions_intersec.append(react)
        masses = {}
        for key in result.keys():
            if key == 'charge':
                charge_intersec.append(result['charge'])
            else:
                masses[key] = result[key]
        if 'charge' not in result.keys():
            charge_intersec.append('')
        mass_intersec.append(masses)
print(len(reactions_intersec), len(charge_intersec), len(mass_intersec))
```

```python
mass_charge_dict = {'Reactions': reactions_intersec, 'Charge': charge_intersec, 'Masses': mass_intersec}
mass_charge = pd.DataFrame(mass_charge_dict)
mass_charge.to_csv('imbalances_intersec.csv')
mass_charge.head(10)
```
